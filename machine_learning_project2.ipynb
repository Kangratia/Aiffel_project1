# 1. 프로젝트 (1) load_digits : 손글씨를 분류해 봅시다

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
```

```python
digits = load_digits()
```

```python
digits_df["label"] = digits.target
digits_df
```

```python
digits_df['label'].value_counts() #balanced dataset
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=42)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

```python
from sklearn.tree import DecisionTreeClassifier 
model_dt = DecisionTreeClassifier(random_state=32) 
model_dt.fit(X_train, y_train)
```

```python
y_pred = model_dt.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier 

X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=42)

random_forest = RandomForestClassifier(random_state=42) 
random_forest.fit(X_train, y_train) # 훈련
y_pred = random_forest.predict(X_test) # 예측

print(classification_report(y_test, y_pred)) # 결과 지표를 확인
```

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```

```python
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=42)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

```python
y_pred = model_wine.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, roc_auc_score

# 모델 예측
predictions = model_wine.predict(X_test)

# 평가 지표 계산
accuracy = accuracy_score(y_test, predictions)
f1 = f1_score(y_test, predictions, average="weighted")
precision = precision_score(y_test, predictions, average="weighted")
recall = recall_score(y_test, predictions, average="weighted")

# 오차 행렬
confusion_matrix = confusion_matrix(y_test, predictions)

# 결과 출력
print("Accuracy:", accuracy)
print("F1 Score:", f1)
print("Precision:", precision)
print("Recall:", recall)
print("Confusion Matrix:", confusion_matrix)
```
